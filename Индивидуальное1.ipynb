```python
class Track:
    def __init__(self, name, duration):
        self.name = name
        self.duration = duration
        if not isinstance(duration, int):
          raise TypeError("Время должно быть числом")
    def show(self):
      if self.duration == 1:
        print(f'{self.name} - {self.duration} минутa')
      if 1 < self.duration < 5:
        print(f'{self.name} - {self.duration} минуты')
      if self.duration > 4:
        print(f'{self.name} - {self.duration} минут')

class Album:
    def __init__(self, name, band):
        self.name = name
        self.band = band
        self.tracks = []

    def get_tracks(self):
        for track in self.tracks:
            track.show()
    def add_track(self, name, duration):
        self.tracks.append(Track(name, duration))
    def get_duration(self):
        total_duration = 0
        for track in self.tracks:
            total_duration += track.duration
        if total_duration == 1:    
          print(f'Длительность альбома "{self.name}" - {total_duration} минутa')
        if 1 < total_duration < 5:    
          print(f'Длительность альбома "{self.name}" - {total_duration} минуты')
        if total_duration > 4:    
          print(f'Длительность альбома "{self.name}" - {total_duration} минут')
    def del_track(self, name):
        for track in self.tracks:
            if track.name == name:
                self.tracks.remove(track)
                print(f'Трек "{name}" удален из альбома "{self.name}"')
                return
        print(f'Трек "{name}" не найден в альбоме "{self.name}"')


album1 = Album('Альбом 1', 'Группа 1')
album1.add_track('Трек 1', 3)
album1.add_track('Трек 2', 3)
album1.add_track('Трек 3', 4)

album2 = Album('Альбом 2', 'Группа 2')
album2.add_track('Трек 1', 2)
album2.add_track('Трек 2', 1)
album2.add_track('Трек 3', 2)

print('Треки альбома 1:')
album1.get_tracks()
print('')
print('Треки альбома 2:')
album2.get_tracks()
print('')
album1.del_track('Трек 2')
print('Треки Альбома 1 после удаления:')
album1.get_tracks()
print('')
album1.get_duration() # 10
album2.get_duration() # 5
```
