# 9.2.7

Реализуйте модуль для numbers.py, позволяющий:

находить НОД и НОК;

определять, является ли число простым;

получать обратное число;

получать корень из числа.

```python
try:
    first = int(input("Введите первое число:"))
    second = int(input("Введите второе число:"))

except TypeError:
    print("Введите целые числа!")
except ValueError:
    print("Введите числа отличные от нуля!")

def gcd(first, second):
    """Вернуть НОД для целых чисел 'first' и 'second'.

    Пример:
        - gcd(54, 24) == 6;
        - gcd(-54, 24) == 6.

    Исключения:
        - TypeError: 'first' или 'second' - любой тип кроме int;
        - ValueError: 'first' или second равны 0.
    """

    while first != 0 and second != 0:
        if first > second:
            first = first % second
        else:
            second = second % first

    nod = first + second
    return nod

print("Наибольший общий делитель:", gcd(first, second))

def lcm(first, second):
    """Вернуть НОК для чисел 'first' и 'second'.

    Пример:
        - lcm(4, 6) == 12;
        - lcm(-4, 6) == 12.

    Исключения:
        - TypeError: 'first' или 'second' - любой тип кроме int;
        - ValueError: 'first' или second равны 0.

    """
    if first > second:
        greater = first
    else:
        greater = second
    while (True):
        if ((greater % first == 0) and (greater % second == 0)):
            lcm1 = greater
            break
        greater += 1
    return lcm1

print("Наименьшее общее кратное:", lcm(first, second))


try:
    number = int(input("Введите число: "))
except TypeError:
    print("Введите целое число!")
except ValueError:
    print("Не натуральное число!")

def is_prime(number):
    """Вернуть True, если 'number' - простое число, иначе False.

    Пример: is_prime(7) == True.

    Исключения:
        - TypeError: 'number' - любой тип кроме int;
        - ValueError: 'number' - не натуральное число.
    """
    k = 0
    for i in range(2, number // 2 + 1):
        if (number % i == 0):
            k = k + 1
    if (k <= 0):
        return True
    else:
        return False

print("Является ли число простым?", is_prime(number))



def inverse(number):
    """Вернуть число, обратное 'number' (1 / number).

    Пример: inverse(2) == 0.5.

    Исключения:
        - TypeError: 'number' - любой тип кроме int, float;
        - ZeroDivisionError: 'number' - 0.
    """
    nm = 1 / number
    return nm

print("Число, обратное к данному", inverse(number))


try:
    power = int(input("Введите степень корня:"))

except TypeError:
    print("Введите число!")
except ValueError:
    print("Введите число, отличное от нуля!")
except ValueError:
    print("Значение степени меньше единицы!")

def root(number, power):
    """Вернуть корень 'power'-степени из 'number'.

    Пример: root(9) == 3.0.

    Исключения:
        - TypeError: 'number' - любой тип кроме int, float;
        - TypeError: 'power' - любой тип кроме int, float;
        - ValueError: 'power' - меньше единицы.
    """
    res = number ** (1 / power)
    return res
print("Корень из числа:", root(number, power))
```
