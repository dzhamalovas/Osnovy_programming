 Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий делитель).

```python
def gcd(first, second):           #gcd - greatest common divisor
  if first > second:
    a = second
  else:
    a = first
  for i in range(1,a+1):
    if (first % i == 0) and (second % i == 0 ):
      gcd = i
  return gcd

def lcm(first, second):          #lcm - Lowest Common Multiple
  if first>second:
    n=first
  else:
    n=second
  while(True):
    if (n%first==0) and (n%second==0):
      return n
    n +=1

def gcd_nums(nums):           #объявление фунциии
  for j in range(1,len(nums)-1): 
    nums[j]=gcd(nums[j-1],nums[j])
  return nums[j] 

def lcm_nums(nums1):    #объявление фунциии
  for j in range(1,len(nums1)):
    nums1[j]=lcm(nums1[j-1],nums1[j])
  return nums1[j]

nums =input("Введите числа: ").split()
nums=list(map(int,nums))
nums1=nums.copy()
nod=gcd_nums(nums)
nok=lcm_nums(nums1)
print("НОД =",nod)
print("НОК =",nok)

# --------------
# Пример вывода:
#
# Введите числа через пробел: 8 10 14
# НОД = 2
# НОК = 280
#
# Введите числа через пробел: 6 8 24 16
# НОД = 2
# НОК = 48

```

```python
def split_numbers(*fun):  #* - неограниченное число аргументов fun
  plus=[]
  minus=[]
  for i in fun:
    if i >= 0:
      plus.append(i)
      plus1=sorted(plus)
    elif i < 0:
      minus.append(i)
  return minus, plus1
  
print(split_numbers(1, 4, -5, 0, -33))

# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])

```
