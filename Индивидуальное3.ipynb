## Задача 1 

```python
class Product:
  def __init__(self, name, price, weight):
    self.name = name
    self.price = price
    self.weight = weight

class Buy(Product):
  def __init__(self, name, price, weight, quantity):
    super().__init__(name, price, weight)
    self.quantity = quantity
    self.total_price = price * quantity
    self.total_weight = weight * quantity

class Check(Buy):
  def show_inf(self):
    print(f"Наименование продукта: {self.name}")
    print(f"Цена за продукт: {self.price}")
    print(f"Вес продукта: {self.weight}")
    print(f"Количество покупаемого товара: {self.quantity}")
    print(f"Стоимость всего купленного товара: {self.total_price}")
    print(f"Вес всех товаров: {self.total_weight}")
    
goods = [
    Check('Хлеб', 49, 450, 1),
    Check('Яблоки', 60, 800, 2),
    Check('Овощная смесь',140, 400, 1),
    Check('Пельмени', 450, 700, 3),
    Check('Макароны', 95, 450, 2),
]

for good in goods:
  good.show_inf()
```

## Задача 2

```python
import math

class Calculator:
    def __init__(self, x):
        self.x = x
    def add(self, y):
        self.x += y
    def subtract(self, y):
        self.x -= y
    def multiply(self, y):
        self.x *= y
    def divide(self, y):
        self.x /= y
    def square_root(self):
        self.x = math.sqrt(self.x)
    def trig_function(self, func):
      self.e = math.radians(self.x)
      print('Число в радианах: ', self.e)
      if func == 'sin':
           self.x = math.sin(self.x)
      elif func == 'cos':
          self.x = math.cos(self.x)


class MemoryCalculator(Calculator):
    def __init__(self, x):
        super().__init__(x)
        self.memory = 0
    
    def write_to_memory(self):
        self.memory = self.x
    
    def read_from_memory(self):
        self.x = self.memory
    
    def add_to_memory(self):
        self.memory += self.x

mc = MemoryCalculator(10)
mc.add(5)
mc.write_to_memory()
mc.subtract(3)
mc.add_to_memory()
mc.read_from_memory()
mc.multiply(2)
print(mc.x) 
```
